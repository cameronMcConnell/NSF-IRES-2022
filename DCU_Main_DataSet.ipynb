```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Create lists of both continuous and classifcation variable indexes

list1 = [x for x in range(100,148)]
list2 = [x for x in range(1,10)]
list3 = [x for x in range(148,159)]
list4 = [x for x in range(462,489)]
list5 = [x for x in range(22,37)]
list6 = [x for x in range(77,91)]
list7 = [x for x in range(40,62)]
list8 = [x for x in range(63,68)]
    
list9 = [72,75,76,449,450,451,452,453,15,255,94,19,21,489,490,491,39,62]
final_list = list1 + list2 + list3 + list4 + list5 + list6 + list7 + list8 + list9
final_list.sort()

continuous_list = [x for x in range(492) if x not in final_list]
```

```python
# Open the data set and then split it using the indexes gained from previous cell
# Each list only contains the names of the columns/variables

df = pd.read_csv(r'C:\Users\Cameron\OneDrive\Documents\DCU_Data\MWBW full dataset 2377 participants_anonymised (1).csv')

df_columns = list(df.columns)
classi_named_columns = [df_columns[x] for x in final_list]
contin_named_columns = [df_columns[x] for x in continuous_list]

contin_df = df[contin_named_columns].copy()
classi_df = df[classi_named_columns].copy()
     
contin_df = contin_df.astype('float') 
classi_df = classi_df.astype('float')   
```

```python
# Split the continous variable names into four seperate lists

one_contin_named_columns,two_contin_named_columns,three_contin_named_columns,four_contin_named_columns = [],[],[],[]

for i, n in enumerate(contin_named_columns):
    if i < (len(contin_named_columns)/4):
        one_contin_named_columns.append(n)
    elif i > (len(contin_named_columns)/4) and i < (len(contin_named_columns)/2):
        two_contin_named_columns.append(n)
    elif i > (len(contin_named_columns)/2) and i < ((len(contin_named_columns)/2) + (len(contin_named_columns)/4)):
        three_contin_named_columns.append(n)
    elif i > ((len(contin_named_columns)/2) + (len(contin_named_columns)/4)):
        four_contin_named_columns.append(n)
```

```python
# Split the classifcation variable names into four seperate lists

one_classi_named_columns,two_classi_named_columns,three_classi_named_columns,four_classi_named_columns = [],[],[],[]

for i, n in enumerate(classi_named_columns):
    if i < (len(classi_named_columns)/4):
        one_classi_named_columns.append(n)
    elif i > (len(classi_named_columns)/4) and i < (len(classi_named_columns)/2):
        two_classi_named_columns.append(n)
    elif i > (len(classi_named_columns)/2) and i < ((len(classi_named_columns)/2) + (len(classi_named_columns)/4)):
        three_classi_named_columns.append(n)
    elif i > ((len(classi_named_columns)/2) + (len(classi_named_columns)/4)):
        four_classi_named_columns.append(n)
```

```python
# Use subplot for the following next four cells
# Create boxplots for each quarter of the four lists of continous variable names
# Export the boxplots to my local directory
# Combine all four pdfs into one single pdf 

fig1, axs1 = plt.subplots(17,5,figsize = (65,65))

n,l = 0,0

for i in one_contin_named_columns:
    
        sns.boxplot(ax = axs1[n,l], x = contin_df[i])
        l += 1
    
        if l == 5:
            l = 0
            n += 1
        
plt.savefig(r'C:\Users\Cameron\OneDrive\Documents\DCU_Data\Contiuous_BoxPlots1.pdf')        
```
