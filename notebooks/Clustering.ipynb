```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import k_means
import pickle
```

```python
cluster_nums = np.arange(start=2, stop=100)
```

```python
entire_matrix = pd.read_pickle('entire_matrix.pickle')
```

```python
costs = np.empty_like(cluster_nums)

for c in cluster_nums:
    centroids, y_hat, cost = k_means(entire_matrix, n_clusters=c)
    costs[c-2] = cost

plt.plot(cluster_nums, costs)
plt.xlabel('Number of Cluster')
plt.ylabel('Cost')

plt.savefig('entire_matrix.png')
```
